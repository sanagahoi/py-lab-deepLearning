```python
import math
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(100)
```

```python
sigmoid(-4)
```

```python
sigmoid(1)
```

```python
def tanh(x):
    return (math.exp(x) - math.exp(-x)) / (math.exp(x) + math.exp(-x))
```

```python
tanh(1)
```

```python
tanh(-56)
```

```python
tanh(50)
```

```python
def relu(X):
    return max(0, X)
```

```python
relu(-5)
```

```python
relu(1)
```

```python
relu(5)
```

```python
def leaky_relu(X):
    return max(0.01 * X, X)
```

```python
leaky_relu(-5)
```

```python
leaky_relu(0)
```

```python
leaky_relu(5)
```
