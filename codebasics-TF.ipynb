```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
X_train_flatten = X_train.reshape(len(X_train), 28*28)
X_train_flatten
```

```python
X_test_flatten = X_test.reshape(len(X_test), 28*28)
X_test_flatten
```

```python
X_test_flatten.shape
```

```python
X_train_flatten.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(10, input_shape=(784,),activation='sigmoid' )
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics = ['accuracy']
              )
model.fit(X_train_flatten, y_train, epochs=5)
```

accuracy isn't great (.27) before we didn't do scaling of X_train and X_test with dividing by 255. so we'll scale the values first before flattening it then see the increased accuracy (.92) 

```python
model.evaluate(X_test_flatten, y_test)
```

```python
model.predict(X_test_flatten)
```

```python
plt.imshow(X_test[0])
```

```python
y_predicted = model.predict(X_test_flatten)
```

```python
y_predicted[0]
```

```python
np.argmax(y_predicted[0]) #to check the highest index of array
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_predicted_labels)
cm
```

```python
import seaborn as sns
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Truth")
```

```python
model_with_hidden_layer = keras.Sequential([
    keras.layers.Dense(100, input_shape=(784,), activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_with_hidden_layer.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model_with_hidden_layer.fit(X_train_flatten, y_train, epochs= 5)
```

```python
model_with_hidden_layer.evaluate(X_test_flatten, y_test)
```

```python
# we don't need to flatten the X_train, and X_test, keras has a feature to do it 

model_keras_flatten = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model_keras_flatten.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model_keras_flatten.fit(X_train, y_train, epochs= 5)
```

```python
# exercise - try different loss functions and optimizers
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(
    optimizer = 'adam',
    loss = keras.losses.SparseCategoricalCrossentropy(),
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs= 5)
```

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (28,28)),
    keras.layers.Dense(100, activation='relu'),
    keras.layers.Dense(10, activation='sigmoid')
])

model.compile(
    optimizer = 'SGD',
    loss = keras.losses.SparseCategoricalCrossentropy(),
    metrics = ['accuracy']
)

model.fit(X_train, y_train, epochs= 5)
```
